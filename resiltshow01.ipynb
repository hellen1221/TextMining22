```python
import pandas as pd

df = pd.read_csv("result01.tsv", sep="\t")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(df['loss'], 'b-', label='loss')
plt.plot(df['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(df['accuracy'], 'g-', label='accuracy')
plt.plot(df['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```
