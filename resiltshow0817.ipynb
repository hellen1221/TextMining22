```python
import pandas as pd

df = pd.read_csv(".\\tdata\\doutN.tsv", sep="\t")
print(df)
```

연습

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(df['loss'], 'b-', label='loss')
plt.plot(df['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(df['accuracy'], 'g-', label='train_accuracy')
plt.plot(df['val_accuracy'], 'k--', label='test_accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.show()
```

```python
Group = ['E1_G6', 'E2_G11', 'E3_G20']
Group_values = [0.918, 0.891, 0.857]

plt.figure(figsize=(7, 5))
plt.bar(Group, Group_values, color='g', alpha=0.6)

for i, name in enumerate(Group):
    plt.text(name, Group_values[i], "{:.3f}".format(Group_values[i]),
             fontsize = 14, 
             color='blue',
             horizontalalignment='center',
             verticalalignment='bottom')

plt.ylim(0.7, 1.0)

plt.show()
```
